```
import matplotlib.pyplot as plt
from collections import Counter
import numpy as np

# =============================================
# 🗑️ Planificador de Recolección de Residuos 🗑️
# =============================================

# ---------------------------------------------
# 📅 Variables con los días de la semana 📅
# ---------------------------------------------
dias_semana = ["Lunes", "Martes", "Miércoles", "Jueves", "Viernes", "Sábado", "Domingo"]

# ---------------------------------------------
# ♻️ Tipos de residuos que se pueden recolectar ♻️
# ---------------------------------------------
tipo_residuo = ["⚠️ Peligroso", "🚫 No_aprovechable", "♻️ Inorgánico", "🍎 Orgánico", "📱 Electrónico"]

# ---------------------------------------------
# 📝 Plan de recolección inicializado 📝
# ---------------------------------------------
plan_recoleccion = {dia: [] for dia in dias_semana}

# ---------------------------------------------
# ➕ Función para agregar recolección a un día específico ➕
# ---------------------------------------------
def agregar_recoleccion():
    print("\n" + "=" * 50)
    print("📅 AGREGAR RECOLECCIÓN 📅".center(50))
    print("=" * 50)
    while True:
        print("\n📅 Días de la semana disponibles:")
        print(", ".join(dias_semana))
        dia_ingresado = input("Ingrese el día de la semana (o 'fin' para terminar): ").capitalize()
        if dia_ingresado == 'Fin':
            break
        if dia_ingresado in plan_recoleccion:
            while True:
                print("\n♻️ Tipos de residuos disponibles:")
                print(", ".join(tipo_residuo))
                tipo_ingresado = input(f"Ingrese el tipo de residuo para el día {dia_ingresado} (o 'fin' para terminar): ").capitalize()
                if tipo_ingresado == 'Fin':
                    break
                if tipo_ingresado in [tr.split(" ")[-1] for tr in tipo_residuo]:
                    plan_recoleccion[dia_ingresado].append(tipo_ingresado)
                    print(f"✅ Se ha agregado {tipo_ingresado} al día {dia_ingresado}.")
                else:
                    print("❌ Tipo de residuo inválido. Por favor, ingrese un tipo válido.")
        else:
            print("❌ Día de la semana inválido. Por favor, ingrese un día válido.")

# ---------------------------------------------
# 📋 Función para mostrar el plan de recolección 📋
# ---------------------------------------------
def mostrar_plan():
    print("\n" + "=" * 50)
    print("📋 PLAN DE RECOLECCIÓN DE RESIDUOS 📋".center(50))
    print("=" * 50)
    for dia, residuos in plan_recoleccion.items():
        if residuos:
            print(f"{dia}:")
            for residuo in residuos:
                print(f"  - {residuo}")
        else:
            print(f"{dia}: No hay recolección programada.")

# ---------------------------------------------
# 🗑️ Función para eliminar recolección de un día específico 🗑️
# ---------------------------------------------
def eliminar_recoleccion():
    print("\n" + "=" * 50)
    print("🗑️ ELIMINAR RECOLECCIÓN 🗑️".center(50))
    print("=" * 50)
    while True:
        print("\n📅 Días de la semana disponibles:")
        print(", ".join(dias_semana))
        dia_ingresado = input("Ingrese el día de la semana (o 'fin' para terminar): ").capitalize()
        if dia_ingresado == 'Fin':
            break
        if dia_ingresado in plan_recoleccion:
            while True:
                print(f"\n♻️ Residuos disponibles en el día {dia_ingresado}:")
                for residuo in plan_recoleccion[dia_ingresado]:
                    print(f"  - {residuo}")
                tipo_ingresado = input("Ingrese el tipo de residuo que desea eliminar (o 'fin' para terminar): ").capitalize()
                if tipo_ingresado == 'Fin':
                    break
                if tipo_ingresado in plan_recoleccion[dia_ingresado]:
                    plan_recoleccion[dia_ingresado].remove(tipo_ingresado)
                    print(f"✅ Se ha eliminado {tipo_ingresado} del día {dia_ingresado}.")
                else:
                    print(f"❌ No se encontró {tipo_ingresado} en el día {dia_ingresado}.")
        else:
            print("❌ Día de la semana inválido. Por favor, ingrese un día válido.")

# ---------------------------------------------
# 📊 Funciones para graficar 📊
# ---------------------------------------------

# ---------------------------------------------------------------------------------
# 📊 Funcion para graficar la Recolección de residuos por tipo y día (Apilado) 📊
# ---------------------------------------------------------------------------------
def graficar_barras_apiladas(plan):
    residuos = [r.split(" ")[-1] for r in tipo_residuo]
    colores = plt.colormaps.get_cmap('Set3')
    datos = {residuo: [] for residuo in residuos}
    for dia in dias_semana:
        conteo = Counter(plan[dia])
        for residuo in residuos:
            datos[residuo].append(conteo.get(residuo, 0))

    bottom = np.zeros(len(dias_semana))
    plt.figure(figsize=(12, 7))
    for idx, residuo in enumerate(residuos):
        plt.bar(dias_semana, datos[residuo], bottom=bottom, label=residuo, color=colores(idx / len(residuos)))
        bottom += np.array(datos[residuo])

    plt.title("Recolección de residuos por tipo y día (Apilado)", fontsize=16, fontweight='bold')
    plt.xlabel("Día de la semana", fontsize=12)
    plt.ylabel("Cantidad de residuos", fontsize=12)
    plt.xticks(rotation=45)
    plt.legend(title="Tipo de residuo")
    plt.tight_layout()
    plt.show()

# -----------------------------------------------------------------------------------------
# 📊 Funcion para graficar la Proporción de tipos de residuos recolectados en la semana 📊
# -----------------------------------------------------------------------------------------
def graficar_proporcion_residuos(plan):
    todos_los_residuos = sum(plan.values(), [])
    conteo = Counter(todos_los_residuos)
    if conteo:
        etiquetas = list(conteo.keys())
        valores = list(conteo.values())
        colores = plt.colormaps.get_cmap('Pastel1')

        plt.figure(figsize=(8, 8))
        wedges, texts, autotexts = plt.pie(
            valores,
            labels=etiquetas,
            autopct='%1.1f%%',
            startangle=140,
            colors=[colores(i / len(etiquetas)) for i in range(len(etiquetas))],
            textprops={'fontsize': 12}
        )

        plt.setp(autotexts, size=13, weight='bold', color='black')
        plt.title("Proporción de tipos de residuos recolectados en la semana", fontsize=16, fontweight='bold')
        plt.axis('equal')
        plt.tight_layout()
        plt.show()
    else:
        print("⚠️ Aún no hay residuos agregados para graficar.")

def mostrar_graficas():
    print("\n📊 Mostrando gráficas de recolección...")
    graficar_barras_apiladas(plan_recoleccion)
    graficar_proporcion_residuos(plan_recoleccion)

# ---------------------------------------------
# 📌 Menú principal usando un ciclo while y condicionales 📌
# ---------------------------------------------
while True:
    print("\n" + "=" * 50)
    print("📌 MENÚ DEL PLANIFICADOR DE RECOLECCIÓN 📌".center(50))
    print("=" * 50)
    print("1. ➕ Agregar recolección para un día")
    print("2. 📋 Mostrar plan semanal")
    print("3. 🗑️ Eliminar recolección de un día")
    print("4. 📊 Ver gráficas de recolección")
    print("5. 🚪 Salir")

    opcion = input("Seleccione una opción (1-5): ")

    if opcion == '1':
        agregar_recoleccion()
    elif opcion == '2':
        mostrar_plan()
    elif opcion == '3':
        eliminar_recoleccion()
    elif opcion == '4':
        mostrar_graficas()
    elif opcion == '5':
        print("🚪 Saliendo del planificador de recolección. ¡Hasta luego!")
        break
    else:
        print("❌ Opción inválida. Por favor, seleccione una opción válida.")

# ---------------------------------------------
# 🏁 Fin del código 🏁
# ---------------------------------------------
```
