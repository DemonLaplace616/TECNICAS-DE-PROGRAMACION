```
# Variables con los d√≠as de la semana
dias_semana = ["Lunes", "Martes", "Mi√©rcoles", "Jueves", "Viernes", "S√°bado", "Domingo"]
# Se definen los tipos de residuos que se pueden recolectar.
tipo_residuo = ["Peligroso", "No aprovechable", "Inorganico", "Org√°nico", "Electr√≥nico"]
plan_recoleccion = {}
for dia in dias_semana:
    plan_recoleccion[dia] = []
# Agregar recolecci√≥n a un d√≠a espec√≠fico
def agregar_recoleccion():
    """Permite al usuario agregar un tipo de residuo a un d√≠a espec√≠fico."""
    while True:
        print("\nDias de la semana: ")
        print("Lunes, Martes, Mi√©rcoles, Jueves, Viernes, S√°bado, Domingo")
        dia_ingresado = input("Ingrese el d√≠a de la semana (o 'fin' para terminar): ").capitalize()
        if dia_ingresado == 'Fin':
            break
        if dia_ingresado in plan_recoleccion:
            while True:
                print("\nTipos de residuos disponibles: ")
                print("Pl√°stico, Vidrio, Papel, Org√°nico, Electr√≥nico")
                tipo_ingresado = input(f"Ingrese el tipo de residuo para el dia {dia_ingresado} (o 'fin' para terminar): ").capitalize()
                if tipo_ingresado == 'Fin':
                    break
                if tipo_ingresado in tipo_residuo:
                    plan_recoleccion[dia_ingresado].append(tipo_ingresado)
                    print(f"Se ha agregado {tipo_ingresado} al d√≠a {dia_ingresado}.")
                else:
                    print("Tipo de residuo inv√°lido. Por favor, ingrese un tipo v√°lido.")
        else:
            print("D√≠a de la semana inv√°lido. Por favor, ingrese un d√≠a v√°lido.")
# Men√∫ principal usando un ciclo while y condicionales
def mostrar_plan():
    """Muestra el plan de recolecci√≥n de residuos."""
    print("\nPlan de recolecci√≥n de residuos:")
    for dia, residuos in plan_recoleccion.items():
        if residuos:
            print(f"{dia}:")
            for residuo in residuos:
                print(f"- {residuo}")
        else:
            print(f"-> {dia}: \tNo hay recolecci√≥n programada.")
# Eliminar recolecci√≥n de un d√≠a espec√≠fico
def eliminar_recoleccion():
    """Permite al usuario eliminar un tipo de residuo de un d√≠a espec√≠fico."""
    while True:
        print("\nDias de la semana: ")
        print("Lunes, Martes, Mi√©rcoles, Jueves, Viernes, S√°bado, Domingo")
        dia_ingresado = input("Ingrese el d√≠a de la semana (o 'fin' para terminar): ").capitalize()
        if dia_ingresado == 'Fin':
            break
        if dia_ingresado in plan_recoleccion:
            while True:
                print(f"Residuos disponibles en el dia {dia_ingresado}: ")
                for residuo in plan_recoleccion[dia_ingresado]:
                    print(f"- {residuo}")
                tipo_ingresado = input("Ingrese el tipo de residuo que desea eliminar (o 'fin' para terminar): ").capitalize()
                if tipo_ingresado == 'Fin':
                    break
                if tipo_ingresado in plan_recoleccion[dia_ingresado]:
                    plan_recoleccion[dia_ingresado].remove(tipo_ingresado)
                    print(f"Se ha eliminado {tipo_ingresado} del d√≠a {dia_ingresado}.")
                else:
                    print(f"No se encontr√≥ {tipo_ingresado} en el d√≠a {dia_ingresado}.")
        else:
            print("D√≠a de la semana inv√°lido. Por favor, ingrese un d√≠a v√°lido.")
# Men√∫ principal usando un ciclo while y condicionales
while True:
    print("\n--- Men√∫ del Planificador de Recolecci√≥n ---")
    print("1. Agregar recolecci√≥n para un d√≠a")
    print("2. Mostrar plan semanal")
    print("3. Eliminar recolecci√≥n de un d√≠a")
    print("4. Salir")

    opcion = input("Seleccione una opci√≥n (1-4): ")

    if opcion == '1':
        agregar_recoleccion()
    elif opcion == '2':
        mostrar_plan()
    elif opcion == '3':
        eliminar_recoleccion()
    elif opcion == '4':
        print("Saliendo del planificador de recolecci√≥n.")
        break
    else:
        print("Opci√≥n inv√°lida. Por favor, seleccione una opci√≥n v√°lida.")
# Fin del c√≥digo
```

```
import matplotlib.pyplot as plt
from collections import Counter
import numpy as np

# =============================================
# üóëÔ∏è Planificador de Recolecci√≥n de Residuos üóëÔ∏è
# =============================================

# ---------------------------------------------
# üìÖ Variables con los d√≠as de la semana üìÖ
# ---------------------------------------------
dias_semana = ["Lunes", "Martes", "Mi√©rcoles", "Jueves", "Viernes", "S√°bado", "Domingo"]

# ---------------------------------------------
# ‚ôªÔ∏è Tipos de residuos que se pueden recolectar ‚ôªÔ∏è
# ---------------------------------------------
tipo_residuo = ["‚ö†Ô∏è Peligroso", "üö´ No_aprovechable", "‚ôªÔ∏è Inorg√°nico", "üçé Org√°nico", "üì± Electr√≥nico"]

# ---------------------------------------------
# üìù Plan de recolecci√≥n inicializado üìù
# ---------------------------------------------
plan_recoleccion = {dia: [] for dia in dias_semana}

# ---------------------------------------------
# ‚ûï Funci√≥n para agregar recolecci√≥n a un d√≠a espec√≠fico ‚ûï
# ---------------------------------------------
def agregar_recoleccion():
    print("\n" + "=" * 50)
    print("üìÖ AGREGAR RECOLECCI√ìN üìÖ".center(50))
    print("=" * 50)
    while True:
        print("\nüìÖ D√≠as de la semana disponibles:")
        print(", ".join(dias_semana))
        dia_ingresado = input("Ingrese el d√≠a de la semana (o 'fin' para terminar): ").capitalize()
        if dia_ingresado == 'Fin':
            break
        if dia_ingresado in plan_recoleccion:
            while True:
                print("\n‚ôªÔ∏è Tipos de residuos disponibles:")
                print(", ".join(tipo_residuo))
                tipo_ingresado = input(f"Ingrese el tipo de residuo para el d√≠a {dia_ingresado} (o 'fin' para terminar): ").capitalize()
                if tipo_ingresado == 'Fin':
                    break
                if tipo_ingresado in [tr.split(" ")[-1] for tr in tipo_residuo]:
                    plan_recoleccion[dia_ingresado].append(tipo_ingresado)
                    print(f"‚úÖ Se ha agregado {tipo_ingresado} al d√≠a {dia_ingresado}.")
                else:
                    print("‚ùå Tipo de residuo inv√°lido. Por favor, ingrese un tipo v√°lido.")
        else:
            print("‚ùå D√≠a de la semana inv√°lido. Por favor, ingrese un d√≠a v√°lido.")

# ---------------------------------------------
# üìã Funci√≥n para mostrar el plan de recolecci√≥n üìã
# ---------------------------------------------
def mostrar_plan():
    print("\n" + "=" * 50)
    print("üìã PLAN DE RECOLECCI√ìN DE RESIDUOS üìã".center(50))
    print("=" * 50)
    for dia, residuos in plan_recoleccion.items():
        if residuos:
            print(f"{dia}:")
            for residuo in residuos:
                print(f"  - {residuo}")
        else:
            print(f"{dia}: No hay recolecci√≥n programada.")

# ---------------------------------------------
# üóëÔ∏è Funci√≥n para eliminar recolecci√≥n de un d√≠a espec√≠fico üóëÔ∏è
# ---------------------------------------------
def eliminar_recoleccion():
    print("\n" + "=" * 50)
    print("üóëÔ∏è ELIMINAR RECOLECCI√ìN üóëÔ∏è".center(50))
    print("=" * 50)
    while True:
        print("\nüìÖ D√≠as de la semana disponibles:")
        print(", ".join(dias_semana))
        dia_ingresado = input("Ingrese el d√≠a de la semana (o 'fin' para terminar): ").capitalize()
        if dia_ingresado == 'Fin':
            break
        if dia_ingresado in plan_recoleccion:
            while True:
                print(f"\n‚ôªÔ∏è Residuos disponibles en el d√≠a {dia_ingresado}:")
                for residuo in plan_recoleccion[dia_ingresado]:
                    print(f"  - {residuo}")
                tipo_ingresado = input("Ingrese el tipo de residuo que desea eliminar (o 'fin' para terminar): ").capitalize()
                if tipo_ingresado == 'Fin':
                    break
                if tipo_ingresado in plan_recoleccion[dia_ingresado]:
                    plan_recoleccion[dia_ingresado].remove(tipo_ingresado)
                    print(f"‚úÖ Se ha eliminado {tipo_ingresado} del d√≠a {dia_ingresado}.")
                else:
                    print(f"‚ùå No se encontr√≥ {tipo_ingresado} en el d√≠a {dia_ingresado}.")
        else:
            print("‚ùå D√≠a de la semana inv√°lido. Por favor, ingrese un d√≠a v√°lido.")

# ---------------------------------------------
# üìä Funciones para graficar üìä
# ---------------------------------------------

# ---------------------------------------------------------------------------------
# üìä Funcion para graficar la Recolecci√≥n de residuos por tipo y d√≠a (Apilado) üìä
# ---------------------------------------------------------------------------------
def graficar_barras_apiladas(plan):
    residuos = [r.split(" ")[-1] for r in tipo_residuo]
    colores = plt.colormaps.get_cmap('Set3')
    datos = {residuo: [] for residuo in residuos}
    for dia in dias_semana:
        conteo = Counter(plan[dia])
        for residuo in residuos:
            datos[residuo].append(conteo.get(residuo, 0))

    bottom = np.zeros(len(dias_semana))
    plt.figure(figsize=(12, 7))
    for idx, residuo in enumerate(residuos):
        plt.bar(dias_semana, datos[residuo], bottom=bottom, label=residuo, color=colores(idx / len(residuos)))
        bottom += np.array(datos[residuo])

    plt.title("Recolecci√≥n de residuos por tipo y d√≠a (Apilado)", fontsize=16, fontweight='bold')
    plt.xlabel("D√≠a de la semana", fontsize=12)
    plt.ylabel("Cantidad de residuos", fontsize=12)
    plt.xticks(rotation=45)
    plt.legend(title="Tipo de residuo")
    plt.tight_layout()
    plt.show()

# -----------------------------------------------------------------------------------------
# üìä Funcion para graficar la Proporci√≥n de tipos de residuos recolectados en la semana üìä
# -----------------------------------------------------------------------------------------
def graficar_proporcion_residuos(plan):
    todos_los_residuos = sum(plan.values(), [])
    conteo = Counter(todos_los_residuos)
    if conteo:
        etiquetas = list(conteo.keys())
        valores = list(conteo.values())
        colores = plt.colormaps.get_cmap('Pastel1')

        plt.figure(figsize=(8, 8))
        wedges, texts, autotexts = plt.pie(
            valores,
            labels=etiquetas,
            autopct='%1.1f%%',
            startangle=140,
            colors=[colores(i / len(etiquetas)) for i in range(len(etiquetas))],
            textprops={'fontsize': 12}
        )

        plt.setp(autotexts, size=13, weight='bold', color='black')
        plt.title("Proporci√≥n de tipos de residuos recolectados en la semana", fontsize=16, fontweight='bold')
        plt.axis('equal')
        plt.tight_layout()
        plt.show()
    else:
        print("‚ö†Ô∏è A√∫n no hay residuos agregados para graficar.")

def mostrar_graficas():
    print("\nüìä Mostrando gr√°ficas de recolecci√≥n...")
    graficar_barras_apiladas(plan_recoleccion)
    graficar_proporcion_residuos(plan_recoleccion)

# ---------------------------------------------
# üìå Men√∫ principal usando un ciclo while y condicionales üìå
# ---------------------------------------------
while True:
    print("\n" + "=" * 50)
    print("üìå MEN√ö DEL PLANIFICADOR DE RECOLECCI√ìN üìå".center(50))
    print("=" * 50)
    print("1. ‚ûï Agregar recolecci√≥n para un d√≠a")
    print("2. üìã Mostrar plan semanal")
    print("3. üóëÔ∏è Eliminar recolecci√≥n de un d√≠a")
    print("4. üìä Ver gr√°ficas de recolecci√≥n")
    print("5. üö™ Salir")

    opcion = input("Seleccione una opci√≥n (1-5): ")

    if opcion == '1':
        agregar_recoleccion()
    elif opcion == '2':
        mostrar_plan()
    elif opcion == '3':
        eliminar_recoleccion()
    elif opcion == '4':
        mostrar_graficas()
    elif opcion == '5':
        print("üö™ Saliendo del planificador de recolecci√≥n. ¬°Hasta luego!")
        break
    else:
        print("‚ùå Opci√≥n inv√°lida. Por favor, seleccione una opci√≥n v√°lida.")

# ---------------------------------------------
# üèÅ Fin del c√≥digo üèÅ
# ---------------------------------------------
```
